```python
import pandas as pd

df = pd.read_csv('./Dataset/clean_output_overwatch.csv')
```

```python
import re

def reduce_repeated_patterns(text):
    pattern = re.compile(r"(.+?)\1{2,}")
    while True:
        # Search for repeating patterns
        match = pattern.search(text)
        if not match:
            break
        text = text.replace(match.group(0), match.group(1))
    return text
```

```python
df['review'] = df['review'].apply(reduce_repeated_patterns)
```

```python
pos_df = df[df['voted_up'] == True]
neg_df = df[df['voted_up'] == False]
```
